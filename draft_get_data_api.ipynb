```python
import requests

API_KEY = '83710a8f-095c-bcdc-5766-243122e4903c'  # <- Ersetze mit deinem echten API-Key
LAT = 52.5200  # z. B. Berlin
LNG = 13.4050
RADIUS = 5  # Umkreis in km (max. 25)

url = "https://creativecommons.tankerkoenig.de/json/list.php"

params = {
    'lat': LAT,
    'lng': LNG,
    'rad': RADIUS,
    'sort': 'dist',  # dist = nach Entfernung, price = nach Preis
    'type': 'all',   # diesel, e5, e10 oder all
    'apikey': API_KEY
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
else:
    print("Fehler beim Abrufen:", response.status_code, response.text)
```
